```python
str_cols = df.select_dtypes(include="object").columns

# Baştaki/sondaki boşlukları sil
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())

# Birden fazla boşluğu tek boşluğa indir
df[str_cols] = df[str_cols].apply(lambda x: x.str.replace(r"\s+", " ", regex=True))

# "-" veya sadece boşluk içeren değerleri NaN yap
df[str_cols] = df[str_cols].apply(lambda x: x.replace(r"^\s*-\s*$", pd.NA, regex=True))

# Boş stringleri NaN yap
df[str_cols] = df[str_cols].apply(lambda x: x.replace("", pd.NA))

# Kolon isimlerini de temizle
df.columns = df.columns.str.strip().str.replace(r"\s+", " ", regex=True)


df.head(10)
```

```python
df.drop("Адрес", axis=1, inplace=True)

df.head()
```

```python
df["Город"] = df["Поставщик"].str.extract(r"г\.\s*([А-Яа-яA-Za-z\- ]+)")

# Поставщик kolonundan şehir kısmını kaldır
df["Поставщик"] = df["Поставщик"].str.replace(r"\s*г\.\s*[А-Яа-яA-Za-z\- ]+", "", regex=True)

# Temizlik (boşluk vs.)
df["Поставщик"] = df["Поставщик"].str.strip()
df["Город"] = df["Город"].str.strip()

# Kontrol
print(df[["Поставщик", "Город"]].head())
```

```python
cols = [c for c in df.columns if c != "Номер накладной"]

mask_group = (
    df.duplicated(subset=cols, keep=False) &
    df.groupby(cols, dropna=False)["Номер накладной"]
      .transform(lambda x: x.isna().any() and x.notna().any())
)

# SADECE NaN olan satırları seç
mask_nan = mask_group & df["Номер накладной"].isna()

def highlight(row):
    if mask_nan.loc[row.name]:
        return ["background-color: red"] * len(row)
    elif mask_group.loc[row.name]:
        return ["background-color: yellow"] * len(row)
    return [""] * len(row)

df.style.apply(highlight, axis=1)
```
